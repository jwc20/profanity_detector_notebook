```python
import json 
```

```python
# t200_toxic_chat_2023-09-03_11-09.json
# vw_toxic_chat_2023-09-03_11-09.json


with open ('./toxicity_200/t200_toxic_chat_2023-09-03_11-09.json') as f:
    t200 = json.load(f) 

with open ('./vowpal_wabbit/vw_toxic_chat_2023-09-03_11-09.json') as f:
    vw = json.load(f) 
    
print("toxicity-200: ", len(t200))
print("vowpal_wabbit: ", len(vw))


```

```python
# display the elements in the t200 list that are not in the vw list
result = [] 

for i in t200:
    if i not in vw:
        # print(i)
        result.append(i)

print(len(result))
print(result)
        
```

```python
# display the elements in the vw list that are not in the t200 list
result = []

for i in vw:
    if i not in t200:
        # print(i)
        result.append(i)
        
print(len(result))
print(result)
```

```python
import pandas as pd
df_t200 = pd.DataFrame(t200)
df_vw = pd.DataFrame(vw)
```

```python
# Rows present in df_t200 but not in df_vw
print("Rows present in df_t200 but not in df_vw")
print(df_t200[~df_t200.isin(df_vw)].dropna())
```

```python
# Rows present in df_vw but not in df_t200
print("Rows present in df_vw but not in df_t200")
# show only the chat_message column, diplay all rows
# print(df_vw[~df_vw.isin(df_t200)].dropna()['chat_message'])

# write to a txt file 
with open('t200_not_vw.txt', 'w') as f:
    for item in df_vw[~df_vw.isin(df_t200)].dropna()['chat_message']:
        f.write("%s\n" % item)
```
