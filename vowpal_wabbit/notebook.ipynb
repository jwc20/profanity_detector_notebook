```python
import json
from typing import List, Dict, Any
from datetime import datetime
```

```python
with open("../chat_2023-09-03_10-32.json", "r") as f:
    data = json.load(f)
    
    # create an empty toxic_list 

    toxic_list: List[Dict[str, Any]] = []


    # print data iff is_toxic is True
    for chat in data:
        if chat["is_toxic"]:
            print(chat["username"], chat["chat_message"])
            toxic_list.append(chat)
    print(toxic_list)
    
    # export toxic_list to a json file
    filename = f"vw_toxic_chat_{datetime.now().strftime('%Y-%m-%d_%H-%M')}.json"
    with open(filename, "w") as f:
        json.dump(toxic_list, f, indent=4)
        
    

            
```
